```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df
```

```python
df.drop(columns=['id' , 'Unnamed: 32'] , inplace=True)
df
```

```python
df.dtypes
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[: , 1:] , df.iloc[:, 0] ,test_size=0.2)
```

```python
# scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
y_test
```

```python
# label encoding
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test  = encoder.transform(y_test)

y_train
```

NUMPY ARRAYS TO PYTORCH

```python
X_train_tensor = torch.from_numpy(X_train)
y_train_tensor = torch.from_numpy(y_train)
X_test_tensor = torch.from_numpy(X_test)
y_test_tensor = torch.from_numpy(y_test)


print(X_train_tensor.shape)
print(X_test_tensor.shape)
print(y_train_tensor.shape)
print(y_test_tensor.shape)
```

```python
# DEFINING THE MODEL

class My_nn():
  def __init__(self, x):
    self.weights = torch.rand(x.shape[1],1 , dtype=torch.float64 , requires_grad = True)
    self.bias    = torch.zeros(1 , dtype=torch.float32 , requires_grad=True)

  def forward_pass(self, x):
    z = torch.matmul(x,self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss_calculation(self , y_pred , y):
    # Clamp predictions to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

    # Calculate loss
    loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
    return loss
```

```python
# learning parameters
learning_rate = 0.1
epochs = 25
```

```python
# TRAINING PIPELINE
model = My_nn(X_train_tensor)

for epoch in range(epochs):
  # forward pass
  y_pred = model.forward_pass(X_train_tensor)

  # loss calculation
  loss = model.loss_calculation(y_pred, y_train)

  # backward pass
  loss.backward()

  with torch.no_grad():
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

  # zero gradients
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  # print loss in each epoch
  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```

```python
# model evaluation
with torch.no_grad():
  y_pred = model.forward_pass(X_test_tensor)
  y_pred = (y_pred > 0.9).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
