# AUTOGRADS
primarily used to train neural networks by efficiently calculating the gradients of a loss function with respect to the model's parameters.

AIM- to calculate chain derivative efficiently.


```python
import torch
```

```python
# manual computation of gradients by creating a function
# EXAMPLE : y= x^2

def dy_dx(x):
  return 2*x

dy_dx(3)
```

Calculating gradient using pytorch autograds

```python
# EXAMPLE : y= x^2

x = torch.tensor(3.0 , requires_grad=True) #requires_grad true is used to compute the gradient

y = x**2
print(x)
print(y)
```

```python
y.backward()

x.grad
```

```python
# EXAMPLE : y= x^2 , z = sin(y)
x = torch.tensor(3.0 , requires_grad=True)
y =x**2
z = torch.sin(y)

print(x)
print(y)
print(z)
print()

# y.backward()
# x.grad

z.backward()
x.grad
```
