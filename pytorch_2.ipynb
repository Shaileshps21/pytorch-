# AUTOGRADS
primarily used to train neural networks by efficiently calculating the gradients of a loss function with respect to the model's parameters.

AIM- to calculate chain derivative efficiently.


```python
import torch
```

```python
# manual computation of gradients by creating a function
# EXAMPLE : y= x^2

def dy_dx(x):
  return 2*x

dy_dx(3)
```

Calculating gradient using pytorch autograds

```python
# EXAMPLE : y= x^2

x = torch.tensor(3.0 , requires_grad=True) #requires_grad true is used to compute the gradient

y = x**2
print(x)
print(y)
```

```python
y.backward()

x.grad
```

EXAMPLE : to compute the backpropogation grad in case of neural network

```python
# manual creation and gradient calculation
# Inputs
x = torch.tensor(6.7)  # Input feature
y = torch.tensor(0.0)  # True label (binary)
w = torch.tensor(1.0)  # Weight
b = torch.tensor(0.0)  # Bias

print(x)
print(y)
print(w)
print(b)

# Binary cross entropy loss for scaler
def binary_cross_entropy_loss(prediction, target):
    epsilon = 1e-8  # To prevent log(0)
    prediction = torch.clamp(prediction, epsilon, 1 - epsilon)
    return -(target * torch.log(prediction) + (1 - target) * torch.log(1 - prediction))

# Forward pass
z = w * x + b  # Weighted sum (linear part)
y_pred = torch.sigmoid(z)  # Predicted probability

# Compute binary cross-entropy loss
loss = binary_cross_entropy_loss(y_pred, y)
loss

# Derivatives:
# 1. dL/d(y_pred): Loss with respect to the prediction (y_pred)
dloss_dy_pred = (y_pred - y)/(y_pred*(1-y_pred))

# 2. dy_pred/dz: Prediction (y_pred) with respect to z (sigmoid derivative)
dy_pred_dz = y_pred * (1 - y_pred)

# 3. dz/dw and dz/db: z with respect to w and b
dz_dw = x  # dz/dw = x
dz_db = 1  # dz/db = 1 (bias contributes directly to z)

dL_dw = dloss_dy_pred * dy_pred_dz * dz_dw
dL_db = dloss_dy_pred * dy_pred_dz * dz_db

print(f"Manual Gradient of loss w.r.t weight (dw): {dL_dw}")
print(f"Manual Gradient of loss w.r.t bias (db): {dL_db}")
```

```python
# using autograd ---easy peasyy
# creation
x = torch.tensor(6.7)
y = torch.tensor(0.0)
w = torch.tensor(1.0 , requires_grad=True)
b = torch.tensor(0.0 , requires_grad=True)
print(x)
print(y)
print(w)
print(b)

# forward pass
z= w*x + b
y_pred = torch.sigmoid(z)
print(y_pred)

# loss calculation
loss = binary_cross_entropy_loss(y_pred, y)
print(loss)

# backward pass and weight, bias updation
loss.backward()
print(w.grad)
print(b.grad)
```
