# TENSORS

```python
import torch
import numpy as np
print(torch.__version__)
```

```python
if torch.cuda.is_available():
  print("GPU is available")
  print(f"Using GPU: {torch.cuda.get_device_name(0)}")
else:
  print("GPU is not available")
```

CREATING A TENSOR

```python
# using empty
a = torch.empty(2,3)
print(type(a))
print(a)
print()

# using zeros
b = torch.zeros(2,3)
print(type(b))
print(b)
print()

# using ones
c = torch.ones(2,3)
print(type(c))
print(c)
print()

# using seed
d = torch.rand(2,3)
print(type(d))
print(d)
print()
```

```python
# using manual seed--- to give the same random value each time
torch.manual_seed(100)
e = torch.rand(2,3)
print(e)
print()

# using tensor
f = torch.tensor([[1,3,5],[4,2,6]])
print(type(f))
print(f)
print()


# other ways
print("using arange-->" , torch.arange(0,10,2))

print("using linspace-->" , torch.linspace(0,10, 6))

print("using eye-->" , torch.eye(5))

print("using full-->" , torch.full((3,3) ,5))
```
