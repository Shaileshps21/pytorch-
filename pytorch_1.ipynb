CREATING A TENSOR

```python
# using empty
a = torch.empty(2,3)
print(type(a))
print(a)
print()

# using zeros
b = torch.zeros(2,3)
print(type(b))
print(b)
print()

# using ones
c = torch.ones(2,3)
print(type(c))
print(c)
print()

# using seed
d = torch.rand(2,3)
print(type(d))
print(d)
print()
```

```python
# using manual seed--- to give the same random value each time
torch.manual_seed(100)
e = torch.rand(2,3)
print(e)
print()

# using tensor
f = torch.tensor([[1,3,5],[4,2,6]])
print(type(f))
print(f)
print()


# other ways
print("using arange-->" , torch.arange(0,10,2))

print("using linspace-->" , torch.linspace(0,10, 6))

print("using eye-->" , torch.eye(5))

print("using full-->" , torch.full((3,3) ,5))
```

TENSOR SHAPE

```python
X = torch.tensor([[1,3,5],[2,5,6]])
print(X.shape)


y = torch.tensor([[1,3,5,7],[2,4,6,8]])
print(y.shape)
print(y)
```

To create a tensor which looks like someone created earlier

```python
torch.empty_like(y)

torch.zeros_like(y)

torch.ones_like(y)

# torch.rand_like(y) #will create a error
torch.rand_like(y , dtype=torch.float32)
```

TENSOR DATA TYPES

```python
print(y.dtype)

# assign data types
z = torch.tensor([1.0, 2.5, 3.0 , 4.1] , dtype=int)
print(z)

z = torch.tensor([1.0, 2.0, 3.0 , 4.0] , dtype=float)
print(z)

# using to() method to convert data types
z.to(torch.int)
print(z)
```

MATHEMATICAL OPERATIONS

```python
x = torch.rand(2,2)
print(x)


# scalar operation
# addition
print(x+2)
# subtraction
print(x-3)
# multiplication
print(x*2)
# division
print(x/2)
# int division
print(x//2)
```

```python
# element wise operations
a = torch.rand(2,3)
b = torch.rand(2,3)

print(a)
print(b)
print()

# add
print(a+b)
# subtraction
print(a-b)
# mulltiplication
print(a*b)
# division
print(a/b)
```

```python
# reduction operations
e = torch.randint(size=(2,3) , low =0 , high=10 , dtype=torch.float32)
print(e)

# sum
torch.sum(e)
# sum along col
torch.sum(e, dim=0)
# sum along row
torch.sum(e, dim=1)

# mean
torch.mean(e)
# min, max, product, standard deviation
torch.argmax(e)
torch.argmin(e)
```

```python
# matrix multiplication
f = torch.randint(size=(2,3) , low=0 , high= 10 , dtype = torch.float32)
g = torch.randint(size=(3,2) , low=5 , high= 10 , dtype = torch.float32)

print(f)
print(g)
print()
# matrix multiplication
print(torch.matmul(f,g))

# transpose
torch.transpose(f,0,1)

# comparison operator---- same as in other languages

# special operators
torch.sqrt(f)
torch.sigmoid(f)
torch.softmax(f, dim=0)
torch.relu(f)
```

INPLACE OPERATIONS

```python
m = torch.rand(2,3)
n = torch.rand(2,3)

print(m)
print(n)
print()

# if i use m+n to perform the addition then it will generate a new memory to store the result, so we use add_ fn
# print(m.add_(n))

m.sub_(n)

m.relu_()
```

COPYING A TENSOR

```python
a = torch.rand(2,3)
print(a)

b = a
print(b)
print()

a[0][0]=20
print(a)
print(b)

print(id(a))
print(id(b))


c= a.clone()
print(c)
print()

a[0][0]=12
print(a)
print(c)

print(id(a))
print(id(c))
```
